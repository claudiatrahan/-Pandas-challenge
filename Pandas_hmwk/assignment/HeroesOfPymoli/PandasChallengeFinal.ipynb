# Player Count

```python
#Find num of unique SN
players = len(pd_df["SN"].value_counts())
#create new dataframe
players_df = pd.DataFrame({"Total Players": [players]})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calc unique items
unique = len(pd_df["Item ID"].unique())
#calc average price
avg_price = round(pd_df["Price"].mean(),2)
#calc num of purchases
num_purch = len(pd_df["Item ID"])
#Calc total revenue
total_rev = pd_df["Price"].sum()

#Create a summary DF
Summary_df = pd.DataFrame({"Number of Unique Items":[unique], 
                           "Average Price":[avg_price], 
                           "Number of Purchases":[num_purch], 
                           "Total Revenue":[total_rev]})
#format summary DF
Summary_df["Average Price"] = Summary_df["Average Price"].map("${:,.2f}".format)
Summary_df["Total Revenue"] = Summary_df["Total Revenue"].map("${:,.2f}".format)
Summary_df
```

* Create a summary data frame to hold the results

```python
#group by gender
gen = pd_df_2.groupby(["Sex"])
#calc number of players by gender
gen_count = gen["SN"].count()
#calc average purchase price by gender
avg_gen_priceby = gen["Price"].mean()
#calc total price by gender
val_gen_priceby = gen["Price"].sum()

#gather list of unique SN from orig data
unique_sn1 = pd_df_2.drop_duplicates(subset="SN")
#group by gender
unique_sn = unique_sn1.groupby(["Sex"])
#count unique SN by gender
gen_count_unique = unique_sn["SN"].count()

pur_count = gen_count

avg_pur_price1 = (val_gen_priceby /gen_count)
new_avg_pur_price = avg_pur_price1.map("${:,.2f}".format)

tot_pur_val = val_gen_priceby
new_tot_pur_val = tot_pur_val.map("${:,.2f}".format)
avg_tot_purpp = (val_gen_priceby/gen_count_unique)
new_avg_tot_purpp = avg_tot_purpp.map("${:,.2f}".format)

pur_count= pd.DataFrame(pur_count)
pur_count.rename(columns = {'SN':'Purchase Count'}, inplace = True)

#Summary DataFrame completed
Gender_Summary_df = pd.DataFrame(pur_count)
Gender_Summary_df["Purchase Count"] = pur_count
Gender_Summary_df["Average Purchase Price"] = new_avg_pur_price
Gender_Summary_df["Total Purchase Value"] = new_tot_pur_val
Gender_Summary_df["Avg Total Purchase per Person"] = new_avg_tot_purpp


Gender_Summary_df
```

## Age Demographics

* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
#sort values, drop duplicates and get number of unique SNs 
pd_df.sort_values("SN", inplace=True)
pd_df.drop_duplicates(subset="SN", inplace=True)
age_pd_df = len(pd_df)
age_pd_df 

#create bins
age_bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,9999]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pd.cut(pd_df["Age"], age_bins, labels=age_groups).head()

pd_df["Age Groups"] = pd.cut(pd_df["Age"], age_bins, labels=age_groups)

# Create a GroupBy object based upon "View Group"
age_demos_df = pd_df.groupby("Age Groups")
total_count = age_demos_df["Age"].count()

#calculate total count
percent_age = round((total_count/age_pd_df) *100,2)
age_demo_df = pd.DataFrame({"Total Count":total_count, "Percentage of Players": percent_age })
#percent_age
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].map("{:,.2f}%".format)
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Create bins in which to place values based on player age
age_bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,9999]
#Create labels for these bins
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Slice the data and place it into bins
pd.cut(pd_df_3["Age"], age_bins, labels=age_groups).head()

#Place the data series into a new column inside of the Dataframe
pd_df_3["Age Group"] = pd.cut(pd_df_3["Age"], age_bins, labels=age_groups)

age_df = pd_df_3.groupby("Age Group")

#purchase count by age group
age_pur_count = age_df["PID"].count()
#total purchase count by age group
total_count = age_demos_df["Age"].count()

#avg purchase count by age group
age_avg_pur_price = age_df["Price"].mean()
#formatting
age_avg_pur_price = age_avg_pur_price.map("${:,.2f}".format)

#total purchase value by age grop
age_total_pur_val = age_df["Price"].sum()
#formatting
new_age_total_pur_val = age_total_pur_val.map("${:,.2f}".format)

#avg total purchase value by age group
age_avg_total_pur_pp = (age_total_pur_val/total_count)
#formatting
age_avg_total_pur_pp = age_avg_total_pur_pp.map("${:,.2f}".format)

#Summary DataFrame completed
Age_Summary_df = pd.DataFrame(age_pur_count)
Age_Summary_df["Purchase Count"] = age_pur_count
Age_Summary_df["Average Purchase Price"] = age_avg_pur_price
Age_Summary_df["Total Purchase Value"] = new_age_total_pur_val
Age_Summary_df["Avg Total Purchase per Person"] = age_avg_total_pur_pp

Age_Summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group data by SN
ts_df = pd_df_4.groupby("SN")
#number of unique purchases of indivdual SNs
ts_purchases = ts_df.size()
#average purchase price by player(SN)--duplicates treaated as individuals
ts_avg =ts_df["Price"].mean()
#total purchase value by player(SN)
ts_total = ts_df["Price"].sum()

#ind top spender avg = total purchase value divided by number of purchase count
ind_ts_avg = round(ts_total/ts_purchases, 2)
#put data in frame
ts_df2 = pd.DataFrame({"Purchase Count": ts_purchases, "Average Purchase Price": ind_ts_avg.map("${:.2f}".format), "Total Purchase Value": ts_total})
#sort data by frequency of SN purchases 
ts_df3 = ts_df2.sort_values("Total Purchase Value", ascending = False)
#show top 5 spenders 
ts_df3.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_pop = pd_df_4.groupby(['Item ID', 'Product','Price'])
mp_count = most_pop.count()
mp_count.sort_values(by=['PID'], ascending=False, inplace=True)
mp_remix = mp_count.rename(columns={'PID':'Purchase Count'})
del mp_remix['SN']
del mp_remix['Age']
del mp_remix['Sex']
mp_remix.reset_index(inplace=True)
mp_remix.set_index('Item ID', inplace=True)
mp_remix['Total Purchase Value'] = mp_remix['Price'] * mp_remix['Purchase Count']
mp_remix['Total Purchase Value'] = mp_remix['Total Purchase Value'].map("${:,.2f}".format)
mp_remix['Price'] =  mp_remix['Price'].map("${:,.2f}".format)
most_popular = mp_remix
most_popular_items = most_popular.head(5)
most_popular_items
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_prof = pd_df_4.groupby(['Item ID', 'Product','Price'])
mp_count = most_prof.count()
mp_count.sort_values(by=['PID'], ascending=False, inplace=True)
mp_new = mp_count.rename(columns={'PID':'Purchase Count'})

del mp_new['SN']
del mp_new['Age']
del mp_new['Sex']

mp_new.reset_index(inplace=True)
mp_new.set_index('Item ID', inplace=True)
mp_new['Total Purchase Value'] = mp_new['Price'] * mp_new['Purchase Count']
mp_new.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)
mp_new['Total Purchase Value']= mp_new['Total Purchase Value'].map("${:,.2f}".format)
mp_new['Price']= mp_new['Price'].map("${:,.2f}".format)
most_profitable = mp_new.head(5)
most_profitable
```
